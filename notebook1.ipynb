```python
# Columns :[Age,Math Marks ,Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
data
```

```python
#Get the shape of the matrix.
np.shape(data)
```

```python

#Find the average age of students.
np.average(data[:,0])
```

```python
#Extract Math marks of all students.
data[:,1]

```

```python
#Find the highest Science mark.
np.max(data[:,2])
```

```python
#Get details of the student who scored more than 90 in Math.
maxi=data[:,1]>90
result=data[maxi]
result
```

```python

#Increase Math marks of all students by 5.
data[:,1]+5
```

```python
booli=data[:,0]<19
booli
```

```python
data[booli]
```

```python
#Calculate the average marks in each subject (column-wise mean).
np.average(data,axis=0)
```

```python
#Get data of students who scored at least 80 in both subjects.
maths=data[:,1]>80
maths

```

```python

science=data[:,2]>80
science
```

```python
actual=(science & maths)
data[actual]
```

```python

#Replace all Science marks < 75 with 0.
marks=data[:,2]<75
data[marks,2]=0
data
```
